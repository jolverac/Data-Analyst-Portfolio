```python
import pandas as pd
```

```python
# Open the CSV File
df_population_raw = pd.read_csv(r'C:\Users\admin\Desktop\Tutorial Jupyter Notebook\population_total.csv')
```

```python
# Look at the Data
df_population_raw
```

```python
# Drop Null Values
df_population_raw.dropna(inplace=True)
```

```python
# Make the Pivot Table
df_pivot = df_population_raw.pivot(index='year', columns='country', values='population')
```

```python
# Select Countries
df_pivot = df_pivot[['United States', 'India', 'China', 'Indonesia', 'Brazil']]
```

```python
# Import Matplotlib
import matplotlib.pyplot as plt
```

```python
# Line Plot
df_pivot.plot(kind='line', xlabel='Year', ylabel='Population', title='Population(1955-2020)', figsize=(8,4))
```

```python
# Show Pivot Table
df_pivot
```

```python
# Select only 2020
df_pivot_2020 = df_pivot[df_pivot.index.isin([2020])]
```

```python
# Show 2020 Population DataFrame
df_pivot_2020
```

```python
# Transpose Rows to Columns
df_pivot_2020 = df_pivot_2020.T
```

```python
# Bar Plot
df_pivot_2020.plot(kind='bar', color='orange', xlabel='Year', ylabel='Population', title='Population 2020', figsize=(8,4))
```

```python
# Select some Years
df_pivot_sample = df_pivot[df_pivot.index.isin([1980, 1990, 2000, 2010, 2020])]
```

```python
# Grouped Bar Plot
df_pivot_sample.plot(kind='bar', xlabel='Year', ylabel='Population', title='Population 2020', figsize=(8,4))
```

```python
# Change Column's Name
df_pivot_2020.rename(columns={2020:'2020'}, inplace=True)
```

```python
# Piechart
df_pivot_2020.plot(kind='pie', y='2020', title='Population 2020(%)')
```
